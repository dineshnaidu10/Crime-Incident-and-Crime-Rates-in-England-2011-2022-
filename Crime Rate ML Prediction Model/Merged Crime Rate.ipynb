```python
#Import Required Library
import pandas as pd
```

```python
#Load all required datasets into a Pandas DataFrame
df_crimerate = pd.read_csv('2011-2022_crimerate.csv')
df_deprivation = pd.read_csv('2011-2022_deprivation.csv')
df_inflation = pd.read_csv('2011-2022_inflationrate.csv')
df_unemployment = pd.read_csv('2011-2022_unemployment.csv')
```

```python
#Merge all DataFrames into a single DataFrame
df_crimerate = pd.merge(df_crimerate, df_deprivation, on=['Month', 'LSOA code'], how='left')
df_crimerate = pd.merge(df_crimerate, df_inflation, on='Month', how='left')
df_crimerate = pd.merge(df_crimerate, df_unemployment, on='Month', how='left')
df_crimerate.head()
```

```python
#Check presence of null values
null_counts = df_crimerate.isnull().sum()   
null_counts
```

```python
#Drop all rows with null values
df_crimerate = df_crimerate.dropna()
```

```python
#Re-check presence of null values
null_counts = df_crimerate.isnull().sum()   
null_counts
```

```python
#Re-order columns to sort features according to deprivation domains and place target feature at the last column
df_crimerate = df_crimerate[['Month', 'LSOA code', 'Count', 'Population', 'Pop_density', 'IMD Score', 
                            'Income Score', 'IDACI Score', 'IDAOPI Score', 'Employment Score', 'EST Score', 
                            'CYP Score', 'AS Score', 'HDD Score', 'Crime Score', 'BHS Score', 'GB Score', 
                            'WB Score', 'LE Score','Indoors Score', 'Outdoors Score', 'Inflation_rate',
                            'unemployment_rate', 'Crime Rate']]
```

```python
#Export DataFrame to a .csv file
df_crimerate.to_csv('2011-2022_mergedcrimerate.csv', index=False)
```
