```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Benchmarking_results.csv')
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(22, 11))
fig.suptitle('Initial ML Models Performance')

sns.set(style='darkgrid',)
sns.lineplot(ax=axes[0, 0], data=df, x='Sample_size', y='R2-Training', hue='ML', style='ML').set(title='R2-Training')
sns.lineplot(ax=axes[0, 1], data=df, x='Sample_size', y='R2-Testing', hue='ML', style='ML').set(title='R2-Testing')
sns.lineplot(ax=axes[1, 0], data=df, x='Sample_size', y='RMSE', hue='ML', style='ML').set(title='RMSE')
sns.lineplot(ax=axes[1, 1], data=df, x='Sample_size', y='MAE', hue='ML', style='ML').set(title='MAE')

axes[0, 0].set(xlabel=' ', ylabel=' ')
axes[0, 1].set(xlabel=' ', ylabel=' ')
axes[1, 0].set(ylabel=' ')
axes[1, 1].set(ylabel=' ')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Initial ML Models Time Performance')

sns.set(style='darkgrid',)
sns.lineplot(ax=axes[0], data=df, x='Sample_size', y='Training_time (s)', hue='ML', style='ML').set(title='Training Time (s)')
sns.lineplot(ax=axes[1], data=df, x='Sample_size', y='Testing_time (s)', hue='ML', style='ML').set(title='Testing Time (s)')
sns.lineplot(ax=axes[2], data=df, x='Sample_size', y='Total_time (s)', hue='ML', style='ML').set(title='Total Time (s)')

for ax in axes.flat:
    ax.set(ylabel=' ')

plt.tight_layout()
plt.show()
```
