```python
#Import required library
import pandas as pd
```

```python
#Load merged crime incident dataset into a Pandas DataFrame
df = pd.read_csv('2011-2022_street.csv')
```

```python
#Create a new column with a value of 1 for all crime incidents
df['Count'] = 1
```

```python
#Create a new DataFrame with only the required columns
df_crimecount = df[['Month', 'LSOA code', 'Count']]
df_crimecount.head()
```

```python
#Groupby the DataFrame by month, LSOA code and sum of crime incidents
#This will give sum of crime incidents for every LSOA code for each particular month
df_crimecount = df_crimecount.groupby(['Month', 'LSOA code'])['Count'].sum().reset_index()
df_crimecount.head()
```

```python
#Export DataFrame to a .csv file
df_crimecount.to_csv('2011-2022_crimecount.csv', index=False)
```
