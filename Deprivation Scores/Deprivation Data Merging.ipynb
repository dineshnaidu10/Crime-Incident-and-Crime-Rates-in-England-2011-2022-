```python
#import necessary libraries
import pandas as pd
from tqdm import tqdm
```

```python
#Loading downloaded .csv files into Pandas DataFrame
df_2010 = pd.read_csv('2010_deprivation_data.csv')
df_2015 = pd.read_csv('2015_deprivation_data.csv')
df_2019 = pd.read_csv('2019_deprivation_data.csv')
```

```python
#Reorganize column orders for all 3 DataFrames to ensure consistency

df_2010 = df_2010[['LSOA CODE', 'IMD SCORE', 'INCOME SCORE', 'IDACI score', 'IDAOPI score', 
                    'EMPLOYMENT SCORE', 'EDUCATION SKILLS AND TRAINING SCORE', 
                    'Children/Young People Sub-domain Score', 'Skills Sub-domain Score',
                    'HEALTH DEPRIVATION AND DISABILITY SCORE', 'CRIME AND DISORDER SCORE', 
                    'BARRIERS TO HOUSING AND SERVICES SCORE',
                    'Geographical Barriers Sub-domain Score','Wider Barriers Sub-domain Score', 
                    'LIVING ENVIRONMENT SCORE', 'Indoors Sub-domain Score', 
                    'Outdoors Sub-domain Score']]
                    
df_2015 = df_2015[['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score', 
                    'Income Score (rate)', 
                    'Income Deprivation Affecting Children Index (IDACI) Score (rate)',
                    'Income Deprivation Affecting Older People (IDAOPI) Score (rate)', 
                    'Employment Score (rate)', 'Education, Skills and Training Score', 
                    'Children and Young People Sub-domain Score', 'Adult Skills Sub-domain Score', 
                    'Health Deprivation and Disability Score', 'Crime Score', 
                    'Barriers to Housing and Services Score', 
                    'Geographical Barriers Sub-domain Score', 
                    'Wider Barriers Sub-domain Score', 'Living Environment Score',                
                    'Indoors Sub-domain Score', 'Outdoors Sub-domain Score', ]]

df_2019 = df_2019[['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score', 
                    'Income Score (rate)', 
                    'Income Deprivation Affecting Children Index (IDACI) Score (rate)',
                    'Income Deprivation Affecting Older People (IDAOPI) Score (rate)', 
                    'Employment Score (rate)', 'Education, Skills and Training Score', 
                    'Children and Young People Sub-domain Score', 'Adult Skills Sub-domain Score', 
                    'Health Deprivation and Disability Score', 'Crime Score', 
                    'Barriers to Housing and Services Score', 
                    'Geographical Barriers Sub-domain Score', 'Wider Barriers Sub-domain Score', 
                    'Living Environment Score', 'Indoors Sub-domain Score', 
                    'Outdoors Sub-domain Score', ]]
```

```python
df_2010.rename(columns={'LSOA CODE':'LSOA code', 'IMD SCORE':'IMD Score', 'INCOME SCORE':'Income Score',
                        'IDACI score':'IDACI Score', 'IDAOPI score':'IDAOPI Score', 
                        'EMPLOYMENT SCORE':'Employment Score', 
                        'EDUCATION SKILLS AND TRAINING SCORE': 'EST Score', 
                        'Children/Young People Sub-domain Score':'CYP Score',
                        'Skills Sub-domain Score':'AS Score', 
                        'HEALTH DEPRIVATION AND DISABILITY SCORE':'HDD Score',
                        'CRIME AND DISORDER SCORE':'Crime Score', 
                        'BARRIERS TO HOUSING AND SERVICES SCORE':'BHS Score',
                        'Geographical Barriers Sub-domain Score':'GB Score', 
                        'Wider Barriers Sub-domain Score':'WB Score',
                        'LIVING ENVIRONMENT SCORE':'LE Score', 'Indoors Sub-domain Score':'Indoors Score', 
                        'Outdoors Sub-domain Score':'Outdoors Score'}, inplace=True)

df_2015.rename(columns={'LSOA code (2011)':'LSOA code', 'Index of Multiple Deprivation (IMD) Score':'IMD Score',
                        'Income Score (rate)':'Income Score', 
                        'Income Deprivation Affecting Children Index (IDACI) Score (rate)':'IDACI Score',
                        'Income Deprivation Affecting Older People (IDAOPI) Score (rate)':'IDAOPI Score',
                        'Employment Score (rate)':'Employment Score', 
                        'Education, Skills and Training Score':'EST Score',
                        'Children and Young People Sub-domain Score':'CYP Score', 
                        'Adult Skills Sub-domain Score':'AS Score',
                        'Health Deprivation and Disability Score':'HDD Score', 
                        'Barriers to Housing and Services Score':'BHS Score',
                        'Geographical Barriers Sub-domain Score':'GB Score', 
                        'Wider Barriers Sub-domain Score':'WB Score',
                        'Living Environment Score':'LE Score', 'Indoors Sub-domain Score':'Indoors Score', 
                        'Outdoors Sub-domain Score':'Outdoors Score'}, inplace=True)

df_2019.rename(columns={'LSOA code (2011)':'LSOA code', 'Index of Multiple Deprivation (IMD) Score':'IMD Score',
                        'Income Score (rate)':'Income Score', 
                        'Income Deprivation Affecting Children Index (IDACI) Score (rate)':'IDACI Score',
                        'Income Deprivation Affecting Older People (IDAOPI) Score (rate)':'IDAOPI Score',
                        'Employment Score (rate)':'Employment Score', 
                        'Education, Skills and Training Score':'EST Score',
                        'Children and Young People Sub-domain Score':'CYP Score', 
                        'Adult Skills Sub-domain Score':'AS Score',
                        'Health Deprivation and Disability Score':'HDD Score', 
                        'Barriers to Housing and Services Score':'BHS Score',
                        'Geographical Barriers Sub-domain Score':'GB Score', 
                        'Wider Barriers Sub-domain Score':'WB Score',
                        'Living Environment Score':'LE Score', 'Indoors Sub-domain Score':'Indoors Score', 
                        'Outdoors Sub-domain Score':'Outdoors Score'}, inplace=True)
```

```python
#Create lists of years and months for iteration purposes
year_list_2010 = ['2011', '2012', '2013', '2014']
year_list_2015 = ['2015', '2016', '2017', '2018']
year_list_2019 = ['2019','2020', '2021']
year_list_2022 = ['2022']
month_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
month_list_2022 = ['01', '02', '03', '04', '05', '06', '07', '08', '09']
```

```python
#Extend all data from 2011-2014
df_2010_list = []

for year in tqdm(year_list_2010, desc='Loop 1', colour='Green'):
    for month in month_list:
        df_2010['Month'] = year + '-' + month
        df_2010_list.append(df_2010)
        df_2010 = df_2010[['LSOA code', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                            'Employment Score', 'EST Score',
                            'CYP Score', 'AS Score', 'HDD Score', 'Crime Score', 'BHS Score', 
                            'GB Score', 'WB Score', 'LE Score', 'Indoors Score', 'Outdoors Score']]

df_2010 = pd.DataFrame()
df_2010 = pd.concat(df_2010_list)
df_2010 = df_2010[['LSOA code', 'Month', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                    'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                    'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 'Indoors Score', 
                    'Outdoors Score']]
```

```python
#Extend all data from 2015-2018
df_2015_list = []

for year in tqdm(year_list_2015, desc='Loop 1', colour='Green'):
    for month in month_list:
        df_2015['Month'] = year + '-' + month
        df_2015_list.append(df_2015)
        df_2015 = df_2015[['LSOA code', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                            'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                            'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 
                            'Indoors Score', 'Outdoors Score']]

df_2015 = pd.DataFrame()
df_2015 = pd.concat(df_2015_list)
df_2015 = df_2015[['LSOA code', 'Month', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                    'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                    'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 'Indoors Score', 
                    'Outdoors Score']]
```

```python
#Extend all data from 2019-2022
df_2019_list = []

for year in tqdm(year_list_2019, desc='Loop 1', colour='Green'):
    for month in month_list:
        df_2019['Month'] = year + '-' + month
        df_2019_list.append(df_2019)
        df_2019 = df_2019[['LSOA code', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                            'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                            'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 
                            'Indoors Score', 'Outdoors Score']]

for year in tqdm(year_list_2022, desc='Loop 2', colour='Green'):
    for month in month_list_2022:
        df_2019['Month'] = year + '-' + month
        df_2019_list.append(df_2019)
        df_2019 = df_2019[['LSOA code', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                            'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                            'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 
                            'Indoors Score', 'Outdoors Score']]

df_2019 = pd.DataFrame()
df_2019 = pd.concat(df_2019_list)
df_2019 = df_2019[['LSOA code', 'Month', 'IMD Score', 'Income Score', 'IDACI Score', 'IDAOPI Score', 
                    'Employment Score', 'EST Score', 'CYP Score', 'AS Score', 'HDD Score', 
                    'Crime Score', 'BHS Score', 'GB Score', 'WB Score', 'LE Score', 'Indoors Score', 
                    'Outdoors Score']]
```

```python
#Merge all deprivation scores from 2011 - 2022
df_deprivation = pd.concat([df_2010, df_2015, df_2019])
```

```python
#Checking null value presence in the Dataframe
null_counts = df_deprivation.isnull().sum()   #check counts of null values
null_counts
```

```python
#Export merged DataFrame to a .csv file
df_deprivation.to_csv('2011-2022_deprivation.csv', index=False)
```
