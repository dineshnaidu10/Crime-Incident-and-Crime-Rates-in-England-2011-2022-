```python
#import necessary libraries
import os
import pandas as pd
from tqdm import tqdm
import time
```

```python
#getting the file names in the directory
data_files = os.listdir()
data_files
```

```python
#Data collection from 2011-2016
df_list = []

path_of_the_directory = data_files[0]
ext = ('street.csv')
for path, dirc, files in tqdm(os.walk(path_of_the_directory), desc='Loop 1', colour='green'):
    for name in files:
        if name.endswith(ext):
            df_list.append(pd.read_csv(path + '/' + name))

df_street_1 = pd.concat(df_list)

#Drop unwanted column
df_street_1.drop(['Context'], axis=1, inplace=True)

#Drop rows with null values
df_street_1 = df_street_1[df_street_1['Crime ID'].notna()]
df_street_1 = df_street_1[df_street_1['Latitude'].notna()]
df_street_1 = df_street_1[df_street_1['Longitude'].notna()]
df_street_1 = df_street_1[df_street_1['Crime type'].notna()]
df_street_1 = df_street_1[df_street_1['LSOA code'].notna()]

#Drop rows with LSOA code that does not start with 'E'. 'W' stands for LSOA codes in Wales
df_street_1 = df_street_1[df_street_1['LSOA code'].str.match('^E')==True]

#Drop rows with duplicate Crime IDs and keep only the latest one
df_street_1.drop_duplicates(subset='Crime ID', keep='last', inplace=True)

#Drop unwanted columns
df_street_1.drop(['Crime ID'], axis=1, inplace=True)

#Replace null value with generic statement
df_street_1['Last outcome category'].fillna('Status update unavailable', inplace=True)

#Export file to .csv format
df_street_1.to_csv('2011-2016_street.csv', index=False)
```
